```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = (10, 6) 
sns.set_theme(palette='muted', style='whitegrid')
```

```python
dataset = pd.read_csv('Minor2-Project-Dataset.csv')
dataset.head()
```

```python
_ = dataset['PJME_MW'].plot.hist(figsize=(15, 5), bins=200, title='Distribution of PJME Load')
```

```python
dataset.plot('date', 'PJME_MW', figsize = (20, 8), xlabel = 'Date', ylabel = 'PJME_MW', title = 'Electric power - daily')
```

```python
_ = dataset.pivot_table(index=dataset['hour'], 
                     columns='dayofweek', 
                     values='PJME_MW',
                     aggfunc='sum').plot(figsize=(15,4),
                     title='PJM East - Daily Trends')
```

```python
sns.pairplot(dataset.dropna(),
             hue='hour',
             x_vars=['hour','dayofweek', 'year','weekofyear', 'month'],
             y_vars='PJME_MW',
             height=7,
             plot_kws={'alpha':0.15, 'linewidth':0}
            )

plt.show()
```
