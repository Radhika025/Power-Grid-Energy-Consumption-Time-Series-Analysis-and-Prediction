```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
```

```python
dataset = pd.read_csv('Minor2-Project-Dataset.csv')
dataset.head()
```

```python
_ = dataset['PJME_MW'].plot.hist(figsize=(15, 5), bins=200, title='Distribution of PJME Load')
```

```python
plot = dataset.plot(style='.', figsize=(15, 5), title='Entire PJM Load 1998-2001')
```

```python
dataset.plot('date', 'PJME_MW', figsize = (20, 8), xlabel = 'Date', ylabel = 'PJME_MW', title = 'Electric power - daily')
```

```python
only_date = dataset.groupby(['dayofmonth','month', 'year']).size().copy()

only_date.plot(figsize=(15, 5))
plt.show()
```

```python
sns.pairplot(dataset.dropna(),
             hue='hour',
             x_vars=['hour','dayofweek', 'year','weekofyear', 'month'],
             y_vars='PJME_MW',
             height=5,
             plot_kws={'alpha':0.15, 'linewidth':0}
            )
#plt.suptitle('Power Use MW by Hour, Day of Week, Year and Week of Year')
plt.show()
```
